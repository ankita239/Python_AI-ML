**Use Jupyter Notebook to execute following code, try to guess what is happing in this code**

```python
import numpy as np
import matplotlib.pyplot as plt
height = np.random.normal(140,20,1000)
plt.figure(dpi=200)
plt.title("Height Distribution")
plt.xlabel("Height(cm)")
plt.ylabel("Frequency")
plt.hist(height, bins=30, ec="gold")
plt.show()
```

This code uses matplotlib and numpy to represent a set of values as a histogram. Numpy is used in Python to perform mathematical operations in a more efficient and faster way whereas matplotlib is used in python to represent data in an interactive way like histograms, bar graphs, pie-charts,etc.

**What are the differences between jupyter and notepad?**

***Notepad***

- Notepads are used to write python scripts

- They follow a top down approach and executes the whole code linearly or in sequence in a notepad only once

- They are controlled by the powersheell of the system

***Jupyter***

- Jupyter Notebooks are a more interactive way of using python. 

- They are divided in cells and follows a non-linear approach of running the cells which makes the cells or the codes in them independent of each other

- We can also add texts and pictures in between them to make them more practical in real life

- The cells automatically generates a output even if no output is determined in the program which makes them more realiable in real life coding

**Write Down difference between interpreter and compiler**

***Interpreters***

- Interpreters execute the code line-by-line and executes each line immediately

- They are usually slower as compared to compilers since they perform analysis and execution at the runtime itself

- The code is not converted into an executable file before the execution of any line

- Programming languages like python use interpreters to execute the programs in it

***Compilers***

- Compilers execute the whole code only once

- They are faster than the interpreters since they perform analysis and execution only once on a code wholetogether

- An executable file needs to be generated before the execution process takes place

- It checks the error before execution while creating the executable files and hence is good for error detection in the program

- Once compiled, the executable file can be accessed in any system without the requirement of the compiler

- Programming languages like C use compilers to execute the codes written in it. Hence we need generate an executable file before the execution in the powershell. Eg-

                                 gcc -o file.exe file.c 

**What is Data Structure, why we need one in programs?**

   Data Structures in python are define the type the data which is stored. They are used to store, manipulate and perform various operations in the program. Since python is an object-oriented language and everything in python is treated as an object. So when we create variables and a certain type of data structure is stored in it, we also use the methods which are pre-defined in that class only. Therefore data structures define how we can operate on the data.
                                                        
Hence data structures are needed to create references to the class of the type and from those references, we access the methods and functions of that particular class to store, retrive and modify data. They help in efficient data management.
